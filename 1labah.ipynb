Задание 2 из Блокнота.

```python
age = int(input("Введите ваш возраст: "))

if age >= 18:
    print("Доступ разрешен")
elif 14 <= age <= 17:
    print("Доступ ограничен")
else:
    print("Доступ запрещен")
```

Задание 3 из Блокнота.

Задание 4 из Блокнота.

```python
sentence = input("Введите текстовое предложение: ")

repetitions = int(input("Введите количество повторений: "))

count = 0

while count < repetitions:
    print(sentence)
    count += 1
```

Задание 5 из Блокнота.

```python
# объявление функции
def print_fio(name, surname, patronymic):
    initials = surname[0] + name[0] + patronymic[0]
    # выводим результат на экран
    print(initials)

# ввод данных
name = input("Введите имя: ")
surname = input("Введите фамилию: ")
patronymic = input("Введите отчество: ")

# вызов функции
print_fio(name, surname, patronymic)
```
